```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd
import re
import time

options = Options()
options.add_argument("--headless")
service = Service(r"C:\Users\bunyo\OneDrive\Desktop\chromDriver\chromedriver-win64\chromedriver.exe")  # o‘z joylashuvini yoz
driver = webdriver.Chrome(service=service, options=options)

driver.get("https://cwur.org/2025.php")
time.sleep(5)  # sahifa to‘liq yuklansin

html = driver.page_source
soup = BeautifulSoup(html, "html.parser")
table = soup.find("table")
rows = table.find_all("tr")[1:]

records = []
for tr in rows:
    tds = [td.get_text(strip=True) for td in tr.find_all("td")]
    if len(tds) < 9: continue
    m = re.match(r"^(\\d+)", tds[0])
    rank = int(m.group(1)) if m else None
    top_percent = tds[0].replace(str(rank), '').replace('Top', '').strip()
    records.append({
        "World Rank": rank,
        "Top Percent": top_percent,
        "Institution": tds[1],
        "Location": tds[2],
        "National Rank": tds[3],
        "Education Rank": tds[4],
        "Employability Rank": tds[5],
        "Faculty Rank": tds[6],
        "Research Rank": tds[7],
        "Score": tds[8],
    })

df1 = pd.DataFrame(records)
df1.to_csv("cwur_2025_rankings.csv", index=False, encoding="utf-8-sig")
driver.quit()
print("✅ Done:", len(df), "rows")
```
